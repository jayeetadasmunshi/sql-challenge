Create a bar chart of average salary by title.


```python
from sqlalchemy import create_engine
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
connection_str ='postgres://postgres:password@ IPaddress:5432/sql-challenge'


```

```python
engine = create_engine(connection_str)
```

```python
salaries_data = pd.read_sql("SELECT * FROM salaries", con=engine)
salaries_data.head()
```

```python
employee_data = pd.read_sql("SELECT * FROM employees", con=engine)
employee_data.head()
```

```python
combined_data = pd.merge(salaries_data, employee_data, on="emp_no", how="inner")
combined_data.head()
```

```python
titles_data = pd.read_sql("SELECT * FROM titles", con=engine)
titles_data.head()
```

```python
titles_data.rename(columns={
    "title_id" : "emp_title_id"
}, inplace=True)
titles_data
```

```python
combined_data1 = pd.merge(combined_data, titles_data, on="emp_title_id", how="inner")
combined_data1.head()
```

```python
grouped_df = combined_data1.groupby("title").mean()
grouped_df
```

```python
revised_grouped_df = grouped_df.drop(columns = "emp_no")
revised_grouped_df
```

```python
revised_grouped_df = revised_grouped_df.reset_index()
revised_grouped_df
```

```python
# Set x_axis, y_axis & Tick Locations
x_axis = revised_grouped_df["title"]
ticks = np.arange(len(x_axis))
y_axis = revised_grouped_df["salary"]
 
# Create Bar Chart Based on Above Data
plt.bar(x_axis, y_axis, align="center", alpha=0.5, color=["k", "r", "g", "m", "b", "c", "y"])

# Create Ticks for Bar Chart's x_axis
plt.xticks(ticks, x_axis, rotation="vertical")

# Set Labels & Title
plt.ylabel("Salaries ($)")
plt.xlabel("Employee Titles")
plt.title("Average Employee Salary by Title")


# Show plot
plt.show()
```
